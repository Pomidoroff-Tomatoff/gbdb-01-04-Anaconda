### Задание 1
Построить диаграмму

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = list([3.5, 3.8, 4.2, 4.5, 5, 5.5, 7])
```

Линейная диаграмма (график таблицной функции)

```python
plt.plot(x, y)
plt.show()
```

Точечная диаграмма 

```python
plt.scatter(x, y)
plt.show()
```
